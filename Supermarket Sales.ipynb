## Statistical Calculations


### Mean,median,variance,standard deviation,sum,maximum value,minimum value calculations


```python
 #Mean,median,variance,standard deviation,sum,maximum value,minimum value calculations

#Maximum 

max_Unit_price = float(round(df['Unit_price'].max(), 2))
max_Quantity = float(round(df['Quantity'].max(), 2))
max_Tax= float(round(df['Tax'].max(), 2))
max_Total= float(round(df['Total'].max(), 2))
max_cogs= float(round(df['cogs'].max(), 2))
max_gross_margin_percentage= float(round(df['gross_margin_percentage'].max(), 2))
max_gross_income= float(round(df['gross_income'].max(), 2))
max_Rating= float(round(df['Rating'].max(), 2))

#Minimum

min_Unit_price = float(round(df['Unit_price'].min(), 2))
min_Quantity = float(round(df['Quantity'].min(), 2))
min_Tax= float(round(df['Tax'].min(), 2))
min_Total= float(round(df['Total'].min(), 2))
min_cogs= float(round(df['cogs'].min(), 2))
min_gross_margin_percentage= float(round(df['gross_margin_percentage'].min(), 2))
min_gross_income= float(round(df['gross_income'].min(), 2))
min_Rating= float(round(df['Rating'].min(), 2))

#Sum

sum_Unit_price= float(round(df['Unit_price'].sum(), 2))
sum_Quantity= float(round(df['Quantity'].sum(), 2))
sum_Tax= float(round(df['Tax'].sum(), 2))
sum_Total= float(round(df['Total'].sum(), 2))
sum_cogs= float(round(df['cogs'].sum(), 2))
sum_gross_margin_percentage= float(round(df['gross_margin_percentage'].sum(), 2))
sum_gross_income= float(round(df['gross_income'].sum(), 2))
sum_Rating= float(round(df['Rating'].sum(), 2))

#Mean

avg_Unit_price= float(round(df['Unit_price'].mean(), 2))
avg_Quantity= float(round(df['Quantity'].mean(), 2))
avg_Tax= float(round(df['Tax'].mean(), 2))
avg_Total= float(round(df['Total'].mean(), 2))
avg_cogs= float(round(df['cogs'].mean(), 2))
avg_gross_margin_percentage= float(round(df['gross_margin_percentage'].mean(), 2))
avg_gross_income= float(round(df['gross_income'].mean(), 2))
avg_Rating= float(round(df['Rating'].mean(), 2))

#Median

median_Unit_price = float(round(df['Unit_price'].median(), 2))
median_Quantity = float(round(df['Quantity'].median(), 2))
median_Tax= float(round(df['Tax'].median(), 2))
median_Total= float(round(df['Total'].median(), 2))
median_cogs= float(round(df['cogs'].median(), 2))
median_gross_margin_percentage= float(round(df['gross_margin_percentage'].median(), 2))
median_gross_income= float(round(df['gross_income'].median(), 2))
median_Rating= float(round(df['Rating'].median(), 2))

#Variance

variance_Unit_price= float(round(df['Unit_price'].var(), 2))
variance_Quantity = float(round(df['Quantity'].var(), 2))
variance_Tax= float(round(df['Tax'].var(), 2))
variance_Total= float(round(df['Total'].var(), 2))
variance_cogs= float(round(df['cogs'].var(), 2))
variance_gross_margin_percentage= float(round(df['gross_margin_percentage'].var(), 2))
variance_gross_income= float(round(df['gross_income'].var(), 2))
variance_Rating= float(round(Home_and_lifestyle_df['Rating'].var(), 2))

#Standard Deviation

std_Unit_price = float(round(df['Unit_price'].std(), 2))
std_Quantity= float(round(df['Quantity'].std(), 2))
std_Tax= float(round(df['Tax'].std(), 2))
std_Total= float(round(df['Total'].std(), 2))
std_cogs= float(round(df['cogs'].std(), 2))
std_gross_margin_percentage= float(round(df['gross_margin_percentage'].std(), 2))
std_gross_income= float(round(df['gross_income'].std(), 2))
std_Rating= float(round(df['Rating'].std(), 2))


headers = ["Parameter", "Unit_price", "Quantity", "Tax","Total","cogs","gross_margin_percentage","gross_income","Rating"]
 
data = [["Mean", avg_Unit_price,avg_Quantity, avg_Tax,avg_Total,avg_cogs,avg_gross_margin_percentage,avg_gross_income,avg_Rating],
        ["Median", median_Unit_price,median_Quantity,median_Tax,median_Total,median_cogs,median_gross_margin_percentage,median_gross_income,median_Rating],
        ["Variance", variance_Unit_price,variance_Quantity,variance_Tax,variance_Total,variance_cogs,variance_gross_margin_percentage,variance_gross_income,variance_Rating],
        ["StandardDeviation", std_Unit_price,std_Quantity,std_Tax,std_Total,std_cogs,std_gross_margin_percentage,std_gross_income,std_Rating],
        ["Min Value",min_Unit_price,min_Quantity ,min_Tax,min_Total ,min_cogs,min_gross_margin_percentage,min_gross_income,min_Rating],
        ["Max Value",max_Unit_price,max_Quantity,max_Tax,max_Total,max_cogs,max_gross_margin_percentage,max_gross_income,max_Rating],
        ["Sum Value",sum_Unit_price,sum_Quantity,sum_Tax,max_Total,sum_cogs,sum_gross_margin_percentage,sum_gross_income,sum_Rating]]

tabulate.PRESERVE_WHITESPACE = True
tabulate.WIDE_CHARS_MODE = False
print(tabulate(data, headers=headers, tablefmt="grid"))
```

```python
#Pie chart

gender_product_counts = df.groupby(['Gender', 'Product_line']).size().unstack()
fig, axes = plt.subplots(1, len(gender_product_counts), figsize=(15, 5))
fig.suptitle('Product Distribution by Gender')
for idx, (Gender, data) in enumerate(gender_product_counts.iterrows()):
    ax = axes[idx]
    ax.pie(data, labels=data.index, autopct='%1.1f%%', startangle=90,)
    ax.axis('equal')  
    ax.set_title(Gender)
plt.show()
```

```python
#Linegraph

# Group by the "Payment" column and calculate the sum for each group
grouped = df.groupby("Payment").sum()

# Reset the index to make "Payment" a regular column
grouped = grouped.reset_index()

# Plot a line graph for  "Total" over "Payment" groups
grouped.plot(x='Payment', y=['Total'], kind='line', marker='o')

# Add labels and a legend
plt.xlabel('Payment')
plt.ylabel('Total')
plt.title('Total by Payment Type')
plt.legend(['Total'])

# Display the graph
plt.show()
```

 - This is the line graph on the relation between payment methods and the total amount
 
 - From this graph,it is clear that  highest number of payments are done by cash

```python
#Linegraph

# Convert the "Date" column to a datetime object
df['Date'] = pd.to_datetime(df['Date'])

# Extract the month and year from the "Date" column
df['Month'] = df['Date'].dt.to_period('M')

# Group the data by "Month" and calculate the sum of "Total" for each month
monthly_total = df.groupby('Month')['Total'].sum()

# Create a line graph
monthly_total.plot(kind='line', marker='o')

# Customize the graph
plt.xlabel('Month(2019)')
plt.ylabel('Total Amount')
plt.title('Total Amount by Month')
plt.grid()

# Display the graph
plt.show()
```

- This is the line graph that shows how much total amount is gained in sales in ech month

- From this graph,it is clear that in the month of January highest total amounts are gained
